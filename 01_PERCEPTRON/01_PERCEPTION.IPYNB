```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
# Set random seed for reproducibility
dataset=pd.read_csv('./placement.csv')
# Display first few rows
dataset.head()
```

```python
sns.scatterplot(data=dataset,x="cgpa",y="resume_score",hue="placed")
plt.show()
```

```python
X = dataset[['cgpa', 'resume_score']]
y = dataset['placed']
```

```python
import pandas as pd
from sklearn.linear_model import Perceptron
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns

# Example dataset format
# Replace this with your actual dataset loading
# data = pd.read_csv("your_dataset.csv")
# For illustration purposes, assume:
# data = pd.DataFrame({'cgpa': [...], 'resume_score': [...], 'placed': [...]})




# Optional: Plot decision boundary

```

```python
# Feature and label separation
X = dataset[['cgpa', 'resume_score']]
y = dataset['placed']

# Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split for evaluation
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, random_state=42, stratify=y)
```

```python

# Train Perceptron
model = Perceptron(
    eta0=0.1,
    max_iter=1000,
    tol=None,           # To ensure all iterations run
    random_state=42,
    shuffle=True
)
model.fit(X_train, y_train)

# Predict and evaluate
y_pred = model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
import numpy as np
def plot_decision_boundary(clf, X, y):
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 200),
                         np.linspace(y_min, y_max, 200))
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    plt.contourf(xx, yy, Z, alpha=0.3)
    plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k', cmap=plt.cm.coolwarm)
    plt.xlabel("CGPA (scaled)")
    plt.ylabel("Resume Score (scaled)")
    plt.title("Perceptron Decision Boundary")
    plt.show()

plot_decision_boundary(model, X_scaled, y)
```
